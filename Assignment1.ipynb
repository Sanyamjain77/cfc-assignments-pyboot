```python
# Question 3

n1 = int(input(("Enter first number: ")))
n2 = int(input(("Enter second number: ")))

def gcdCal(n1, n2) :
     min= n1;
     gcd = 1;
     if( n2 < n1 ) :
       min = n2
    
     for i in range(2, min):

       while( n1%i == 0 or n2%i == 0 ) :

        if( n1%i == 0 and n2%i == 0 ) :
           gcd *= i
           n1 /= i
           n2 /= i
        elif( n1%i == 0 and n2%i != 0 ) :
           n1 /= i
        else :
           n2 /= i

        if( n1 == 0 or n2 == 0 ) :
          break
    
     return gcd



print(gcdCal(n1, n2))
```

```python
# Question 5

n = int(input(("Enter the number: ")))

count = 0
rem = 0
while n > 0 :
    # cal. the remainder for later
    rem = n % 10
    # eliminating the last digit
    n = n / 10
    n = n - (rem/10)
    # inc. the count of digits
    count = count + 1

print(count);
```

```python
# Question 6

n = int(input(("Enter the number: ")))

temp = n
count = 0
rem = 0
# calculating the no. of digits
while temp > 0 :
    rem = temp%10
    temp = temp / 10
    temp = temp - (rem/10)
    count = count+1

# reversing the no. storing into temporary variable
while count > 0 :
    rem = n%10
    temp = (temp*10) + rem
    n = (n/10) - (rem/10)
    count = count - 1

# number reversed
n = int(temp)

print(n)
```
